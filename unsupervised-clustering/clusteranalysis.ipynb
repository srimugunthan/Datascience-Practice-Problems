```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
clust_data_df = pd.read_excel('../datasets/DAPL - DAT - 1 CLA - eStore.xlsx'
                              , sheet_name='Data')
clust_data_df
```

```python
clust_data_df.columns
```

```python
clustdf = clust_data_df[['BK', 'STA', 'DVD', 'EDV', 'BG', 'HF', 'CLT']]
```

```python
clustdf
```

# kmeans cluster

```python
kmeans = KMeans(3)
kmeans.fit(clustdf)
```

```python
identified_clusters = kmeans.fit_predict(clustdf)
identified_clusters
```

```python
clust_data_df['Clusters'] = identified_clusters 
clustdf['Clusters'] = identified_clusters
```

```python
clustdf
```

```python
clust_data_df
```

```python
plt.scatter(clust_data_df['Case'],clust_data_df['CLT'],c=clust_data_df['Clusters'],cmap='rainbow')
```

# metrics

```python
from sklearn.metrics import silhouette_score
X = clustdf[['BK', 'STA', 'DVD', 'EDV', 'BG', 'HF', 'CLT']]
labels = kmeans.labels_
print("silhouette score", silhouette_score(X, labels, metric='euclidean'))
print("sum of squared distance", kmeans.inertia_)
```

# elbow method

```python
# https://www.analyticsvidhya.com/blog/2021/04/k-means-clustering-simplified-in-python/
```

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(clustdf)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python


number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title("wcss vs number of clusters")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

# PCA

```python
cldf = clust_data_df[['BK', 'STA', 'DVD', 'EDV', 'BG', 'HF', 'CLT']]
```

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components = 2)
 
pca_output_comp = pca.fit_transform(cldf)

 
explained_variance = pca.explained_variance_ratio_
```
