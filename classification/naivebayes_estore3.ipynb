```python
import numpy as np 
import pandas as pd 	
import matplotlib.pyplot as plt 
import math
```

```python

    """
        Bayes Theorem:
                                        Likelihood * Class prior probability
                Posterior Probability = -------------------------------------
                                            Predictor prior probability

                                         P(x|c) * p(c)
                               P(c|x) = ------------------ 
                                              P(x)
    """
```

```python
def accuracy_score(y_true, y_pred):
    """score = (y_true - y_pred) / len(y_true) """
    return round(float(sum(y_pred == y_true))/float(len(y_true)) * 100 ,2)
```

```python
def pre_processing(df):
    """ partioning data into features and target """
    X = df.drop([df.columns[-1]], axis = 1)
    y = df[df.columns[-1]]
    return X, y
```

```python
FEATURES = list
LIKELIHOODS = {}
CLASS_PRIORS = {}
PRED_PRIORS = {}

X_TRAIN = np.array
Y_TRAIN = np.array
TRAIN_SIZE = int
NUM_FEATS = int
```

```python



# df = pd.read_table("./weather.txt")
# #print(df)

# #Split fearures and target
# X,y  = pre_processing(df)


clf_data_df = pd.read_excel('../../analysismethods/DAPL - DAT - 5 NBC - eStore.xlsx',
                                sheet_name='ClassificationData')
clf_data_df

X = clf_data_df[[ 'BK', 'STA', 'DVD', 'EDV', 'BG', 'HF', 'CLT']]
y =   clf_data_df[ 'Cluster']


FEATURES = list(X.columns)
X_TRAIN = X
Y_TRAIN = y
TRAIN_SIZE = X.shape[0]
NUM_FEATS = X.shape[1]

for feature in FEATURES:
    LIKELIHOODS[feature] = {}
    PRED_PRIORS[feature] = {}

    for feat_val in np.unique(X_TRAIN[feature]):
        PRED_PRIORS[feature].update({str(feat_val): 0})

        for outcome in np.unique(Y_TRAIN):
            LIKELIHOODS[feature].update({str(feat_val)+'_'+str(outcome):0})
            CLASS_PRIORS.update({str(outcome): 0})

print(PRED_PRIORS)

print("===========")
print(LIKELIHOODS)

print("===========")
print(CLASS_PRIORS)

print("===========")
```

```python
Y_TRAIN
```

```python
#    def _calc_class_prior(self):
#         """ P(c) - Prior Class Probability """
for outcome in np.unique(Y_TRAIN):
    outcome_count = sum(np.array(Y_TRAIN == outcome))
    CLASS_PRIORS[str(outcome)] = outcome_count / TRAIN_SIZE
    
    
print(PRED_PRIORS)

print("===========")
print(LIKELIHOODS)

print("===========")
print(CLASS_PRIORS)

print("===========")
```

```python
# def _calc_likelihoods(self):
#     """ P(x|c) - Likelihood """
for feature in FEATURES:
    for outcome in np.unique(Y_TRAIN):
        outcome_count = sum(np.array(Y_TRAIN == outcome))
        y_outcome_index = (Y_TRAIN[Y_TRAIN == outcome].index.values.tolist())
        feat_likelihood = X_TRAIN[feature][y_outcome_index].value_counts().to_dict()
        for feat_val, count in feat_likelihood.items():
            LIKELIHOODS[feature][str(feat_val) + '_' + str(outcome)] = count/outcome_count
            
            
print(PRED_PRIORS)

print("===========")
print(LIKELIHOODS)

print("===========")
print(CLASS_PRIORS)

print("===========")
```

```python

#     def _calc_predictor_prior(self):
#         """ P(x) - Evidence """
for feature in FEATURES:
    feat_vals = X_TRAIN[feature].value_counts().to_dict()
    for feat_val, count in feat_vals.items():
        PRED_PRIORS[feature][str(feat_val)] = count/TRAIN_SIZE
        
print(PRED_PRIORS)

print("===========")
print(LIKELIHOODS)

print("===========")
print(CLASS_PRIORS)

print("===========")
```

```python
# def predict( X):
#     """ Calculates Posterior probability P(c|x) """
results = []
X = np.array(X)
for query in X:
    probs_outcome = {}
    for outcome in np.unique(Y_TRAIN):
        prior = CLASS_PRIORS[str(outcome)]
        likelihood = 1
        evidence = 1

        for feat, feat_val in zip(FEATURES, query):
            likelihood *= LIKELIHOODS[str(feat)][str(feat_val) + '_' + str(outcome)]
            evidence *= PRED_PRIORS[str(feat)][str(feat_val)]

        posterior = (likelihood * prior) / (evidence)
        probs_outcome[outcome] = posterior
    result = max(probs_outcome, key = lambda x: probs_outcome[x])
    results.append(result)
print(results)
```

```python
      
print("Train Accuracy: {}".format(accuracy_score(y, np.array(results))))

```
